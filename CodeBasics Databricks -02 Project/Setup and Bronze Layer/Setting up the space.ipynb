```python
import os

volume_path = "/Volumes/project/bronze/original_data"
csv_files = [f for f in os.listdir(volume_path) if f.endswith('.csv')]

for csv_file in csv_files:
    df = spark.read.csv(os.path.join(volume_path, csv_file), header=True, inferSchema=True)
    print(f"Schema for {csv_file}:")
    df.printSchema()
```
