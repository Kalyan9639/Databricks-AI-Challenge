## What do you mean by Parameterized Notebooks?

Parameterized notebooks in Databricks allow you to pass input parameters to a notebook, making your workflows more flexible and reusable. This is especially useful for orchestrating notebook workflows, modularizing code, and automating tasks. The most common way to parameterize notebooks is by using Databricks widgets or by passing arguments when invoking a notebook from another notebook using the `dbutils.notebook.run()` method.

```python
dbutils.widgets.text("widget_name",defaultValue="This is the default value",label="label_name")
```

```python
dbutils.widgets.get("widget_name")
```

```python
dbutils.widgets.text("first",'5')
dbutils.widgets.text("second",'5')
```

```python
dbutils.widgets.dropdown('dropdown','True',['True','False'])
```

```python
dbutils.widgets.get('dropdown')
```

```python
t = int(dbutils.widgets.get('first'))
s = int(dbutils.widgets.get('second'))

dbutils.widgets.multiselect('operators','add',['add','sub','mul','div','pow'])
operator = dbutils.widgets.get("operators")

match operator:
    case 'add': print(t+s)
    case 'sub': print(t-s)
    case 'mul': print(t*s)
    case 'div': print(t/s)
    case 'pow': print(t**s)
```
