```python
import mlflow
import mlflow.spark
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# 1. Setup MLflow Experiment
experiment_path = "/Workspace/Users/kalyanmistcse@gmail.com/Databricks 14 Day AI Challenge - 02/Day - 14"
mlflow.set_experiment(experiment_path)

# 2. Start Manual Run (No Autologging)
with mlflow.start_run(run_name="PlantGrowth_Manual_RF_Optimized") as run:
    
    # 3. Data Preparation
    df = spark.table("dai.phase2.plant_growth_data")
    train, test = df.randomSplit([0.7, 0.3], seed=42)
    
    # 4. Define Pipeline Stages
    soil_indexer = StringIndexer(inputCol="Soil_Type", outputCol="soil_index", handleInvalid="keep")
    water_indexer = StringIndexer(inputCol="Water_Frequency", outputCol="water_index", handleInvalid="keep")
    fert_indexer = StringIndexer(inputCol="Fertilizer_Type", outputCol="fert_index", handleInvalid="keep")
    
    assembler = VectorAssembler(
        inputCols=["soil_index", "water_index", "fert_index", "Sunlight_Hours", "Temperature", "Humidity"],
        outputCol="features"
    )
    
    # 5. Random Forest with Manual Hyperparameters
    rf_trees = 24
    rf_depth = 10
    
    rf = RandomForestClassifier(
        featuresCol="features", 
        labelCol="Growth_Milestone",
        numTrees=rf_trees,
        maxDepth=rf_depth,
        seed=42
    )
    
    # 6. Fit Pipeline
    pipeline = Pipeline(stages=[soil_indexer, water_indexer, fert_indexer, assembler, rf])
    pipeline_model = pipeline.fit(train)
    
    # 7. Evaluation
    predictions = pipeline_model.transform(test)
    evaluator = MulticlassClassificationEvaluator(
        labelCol="Growth_Milestone", 
        predictionCol="prediction", 
        metricName="accuracy"
    )
    accuracy = evaluator.evaluate(predictions)
    
    # 8. MANUAL MLFLOW LOGGING
    # Log Parameters
    mlflow.log_param("model_type", "RandomForestClassifier")
    mlflow.log_param("num_trees", rf_trees)
    mlflow.log_param("max_depth", rf_depth)
    
    # Log Metrics
    mlflow.log_metric("accuracy", accuracy)
    
    # Log Model with Signature
    input_example = train.limit(5).toPandas()
    mlflow.spark.log_model(
        spark_model=pipeline_model, 
        artifact_path="random_forest_model",
        input_example=input_example,
        dfs_tmpdir="/Volumes/dai/phase2/gold/tmp"
    )

    print(f"Manual Run Successful.")
    print(f"Test Accuracy: {accuracy:.2%}")
    print(f"Run ID: {run.info.run_id}")
```

```python
import mlflow
import mlflow.spark
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# 1. Setup MLflow Experiment
experiment_path = "/Workspace/Users/kalyanmistcse@gmail.com/Databricks 14 Day AI Challenge - 02/Day - 14"
mlflow.set_experiment(experiment_path)

with mlflow.start_run(run_name="PlantGrowth_Enhanced_FeatureEng") as run:
    df = spark.table("dai.phase2.plant_growth_data")
    train, test = df.randomSplit([0.7, 0.3], seed=42)
    
    # --- ENHANCED FEATURE ENGINEERING ---
    
    # A. Indexing + One-Hot Encoding (Removes numerical bias from categories)
    s_indexer = StringIndexer(inputCols=["Soil_Type", "Water_Frequency", "Fertilizer_Type"], 
                              outputCols=["s_idx", "w_idx", "f_idx"], handleInvalid="keep")
    
    encoder = OneHotEncoder(inputCols=["s_idx", "w_idx", "f_idx"], 
                            outputCols=["s_vec", "w_vec", "f_vec"])
    
    # B. Assemble all features
    assembler = VectorAssembler(
        inputCols=["s_vec", "w_vec", "f_vec", "Sunlight_Hours", "Temperature", "Humidity"],
        outputCol="unscaled_features"
    )
    
    # C. Standard Scaling (Normalizes the numerical ranges)
    scaler = StandardScaler(inputCol="unscaled_features", outputCol="features")
    
    # D. Tuned Random Forest
    rf = RandomForestClassifier(
        featuresCol="features", 
        labelCol="Growth_Milestone",
        numTrees=24,    # Increased trees for better stability
        maxDepth=12,     # Slightly deeper to find hidden patterns
        subsamplingRate=0.8, # Prevents overfitting on the small dataset
        seed=42
    )
    
    # 2. Fit Pipeline
    pipeline = Pipeline(stages=[s_indexer, encoder, assembler, scaler, rf])
    pipeline_model = pipeline.fit(train)
    
    # 3. Evaluation
    predictions = pipeline_model.transform(test)
    evaluator = MulticlassClassificationEvaluator(labelCol="Growth_Milestone", metricName="accuracy")
    accuracy = evaluator.evaluate(predictions)
    
    # 4. Manual Logging
    mlflow.log_param("engineering", "OneHot+Scaling")
    mlflow.log_metric("accuracy", accuracy)
    
    mlflow.spark.log_model(pipeline_model, "enhanced_model", 
                           dfs_tmpdir="/Volumes/dai/phase2/gold/tmp")

    print(f"Accuracy with Enhanced Engineering: {accuracy:.2%}")
```
